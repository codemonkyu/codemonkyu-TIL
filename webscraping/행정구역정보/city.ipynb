## 광역시도를 아규먼트로 받아서 광역시도중 가장 인구가 많은 행정구역을 반환하는 함수 정의하기

```python
def find_city_max(city):
    pop_max_value = data.loc[data['광역시도'] == city,'인구수'].max()
    return data.loc[(data['인구수'] == pop_max_value) & (data['광역시도'] == city), ['광역시도','행정구역','인구수']]
    
find_city_max('경기도')
```
